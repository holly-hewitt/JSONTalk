```python
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
```

```python


question1_1 = np.array([[18, 0], [8, 14]])
question1_2 = np.array([[16, 1, 1], [14, 2, 2]])
question3_1 = np.array([[13, 6], [12, 8]])
question3_2 = np.array([[12, 3, 4], [13, 1, 6]])

# Define the contingency table
#data = np.array([[a_correct, a_incorrect, a_unsure], 
 #                [b_correct, b_incorrect, b_unsure]])

def perform_test(data, question):
  # Perform the chi-squared test
  stat, p, dof, expected = chi2_contingency(data)

  # Print the results
  print(question)
  print(f"Chi-squared statistic: {stat:.4f}")
  print(f"P-value: {p:.4f}")
  print(f"Degrees of freedom: {dof}")
  print("Expected values:")
  print(expected)

perform_test(question1_1, "\nQuestion 1.1")
perform_test(question1_2, "\nQuestion 1.2")
perform_test(question3_1, "\nQuestion 3.1")
perform_test(question3_2, "\nQuestion 3.2")
```

```python
import scipy.stats as stats
import numpy as np

# Two sets of data, group A and group B
group_A_q1 = [1148.718, 439.938, 140.616, 557.654, 753.073, 491.435, 571.012, 36.248, 10.071, 27.609, 60.317, 64.566, 52.219, 20.102, 26.617, 142.15, 157.951, 77.708]
group_B_q1 = [305.341, 238.832, 252.541, 9.206, 291.01, 251.49, 471.901, 290.69, 179.35, 113.02, 326.931, 132.299, 153.317, 270.18, 540.542, 234.696, 431.685, 783.444, 565.155, 179.864, 369.557, 562.207]

group_A_q2 = [ 359.56,   131.272,  366.719,  217.478,  712.874,  177.482, 1504.587, 1075.848,
  108.996,  205.775,  420.776,  164.601,  214.529,  275.621,  215.031,  164.907,
  194.723,  630.33,   144.982]
group_B_q2 = [570.377, 326.551, 508.286, 200.814, 589.613, 832.243, 362.942, 463.89, 409.964, 178.954, 
                 421.411, 318.377, 288.138, 292.078, 695.66, 233.984, 1232.323, 634.243, 504.752, 429.287,
                 606.361, 217.478]

group_A_q3 = [614.431,  67.667,  18.339,  90.403, 280.145, 267.747, 405.977, 8366.749, 12.442,
 168.799, 227.96,  154.43,   99.993,  69.236, 119.989, 109.37,  102.32,   96.527,
   8.186]
group_B_q3 = [83.006, 79.369, 140.99, 9.342, 199.305, 347.774, 163.845, 138.502, 83.835, 66.552, 116.695, 60.928, 101.872, 137.521, 253.046, 691.443, 174.858, 22182.355, 229.694, 185.205]


def perform_u_test(group_A, group_B, question):
  print("\n"+question)
  # Compute the Mann-Whitney U test statistic and p-value
  u_statistic, p_value = stats.mannwhitneyu(group_A, group_B)

  # Print the results
  print("Mann-Whitney U test statistic:", u_statistic)
  print("P-value:", p_value)

  # Determine whether the result is statistically significant
  if p_value < 0.05:
      print("The result is statistically significant, indicating a significant difference between the two groups.")
  else:
      print("The result is not statistically significant, indicating no significant difference between the two groups.")

  # Compute the effect size using Cohen's d
  n1 = len(group_A)
  n2 = len(group_B)
  ranked_data = stats.rankdata(group_A + group_B)
  mean_rank_A = np.mean(ranked_data[:n1])
  mean_rank_B = np.mean(ranked_data[n1:])
  s = np.std(ranked_data, ddof=1)
  d = (mean_rank_A - mean_rank_B) / s
  print("Effect size (Cohen's d):", d)

perform_u_test(group_A_q1, group_B_q1, "Question 1")
perform_u_test(group_A_q2, group_B_q2, "Question 2")
perform_u_test(group_A_q3, group_B_q3, "Question 3")
 
```
